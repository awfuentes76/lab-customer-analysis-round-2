```python
import pandas as pd
import numpy as np

def transformar_dataframe(df):
    # Mostrar la forma del DataFrame
    print("Forma inicial del DataFrame:", df.shape)
    
    # Eliminar filas con valores NaN
    df = df.dropna()
    print("Forma del DataFrame después de eliminar NaN:", df.shape)
    
    # Estandarizar los nombres de las cabeceras
    df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace(r'[^a-zA-Z0-9_]', '')
    
    # Identificar y mostrar las columnas numéricas y categóricas
    columnas_numericas = df.select_dtypes(include=[np.number]).columns.tolist()
    columnas_categoricas = df.select_dtypes(exclude=[np.number]).columns.tolist()
    print("Columnas numéricas:", columnas_numericas)
    print("Columnas categóricas:", columnas_categoricas)
    
    # Convertir la columna 'effective_to_date' al formato datetime, extraer el mes
    if 'effective_to_date' in df.columns:
        df['effective_to_date'] = pd.to_datetime(df['effective_to_date'])
        df['mes'] = df['effective_to_date'].dt.month
        # Filtrar por el primer trimestre
        df = df[df['mes'].isin([1, 2, 3])]
    else:
        print("La columna 'effective_to_date' no existe en el DataFrame.")
    
    # Retornar el DataFrame transformado
    return df

# Supongamos que tienes un DataFrame llamado df para transformar
# df_transformado = transformar_dataframe(df)
# print("Forma final del DataFrame:", df_transformado.shape)
```
